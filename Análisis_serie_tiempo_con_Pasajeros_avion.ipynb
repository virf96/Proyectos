International Airline Passengers dataset, totales mensuales en miles desde Enero 1949 a Diciembre 1960.

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#index_col='Month', para indexar la fecha
#parse_dates=True, para convertir la variable de fecha a formato fecha
airline = pd.read_csv('https://raw.githubusercontent.com/virf96/curso-series-temporales/master/3.%20Caracter%C3%ADsticas%20fundamentales/Python/airline_passengers.csv',index_col='Month',parse_dates=True)
```

```python
#Observamos una tendencia creciente
#A simple vista podríamos ver una tendencia lineal pero será más facíl observarlo al descomponer la serie 
airline.plot();
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
#Utilizamos el modelo aditivo para descomponer la serie en sus 3 factores.
#Tendencia, Estacionalidad y Residual
#El efecto de tendencia es positivo y nos centraremos en descomponer ese efecto para ver que función describe su comportamiento
#El componente de estacionalidad en general lo que hace es quitarle el componenete de tendencia a la serie original y observamos cierta estacionalidad
#El componente residual explica cualquier tipo de error que no es explicada por los otros 2 factores, también nos dice en donde hay más ruido en la serie
#en este caso, observamos más ruido al principio y al final de la serie 
result = seasonal_decompose(airline['Thousands of Passengers'], model='additive')  
result.plot();
```

### Utilizaremos un modelo aditivo cuando la tendencia tenga un comportamiento más lineal.

### El modelo ultiplicativo es más adecuado cuando la tendencia tenga un comportamiento exponencial o cuando las tasas aumentan o disminuyen a una tasa no lineal

```python
from statsmodels.tsa.seasonal import seasonal_decompose
#Utilizamos el modelo multiplicativo para descomponer la serie en sus 3 factores.

result = seasonal_decompose(airline['Thousands of Passengers'], model='multiplicative')  
result.plot();
```
